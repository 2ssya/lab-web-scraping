```python
# Your solution goes here
# ============================================================
# IMPORTS
# English: We import the three libraries we need:
#   - requests   → makes HTTP GET requests (downloads the HTML page)
#   - BeautifulSoup → parses HTML so we can search and extract data
#   - pandas     → organises the extracted data into a table (DataFrame)
#
# Español: Importamos las tres librerías necesarias:
#   - requests   → hace peticiones HTTP GET (descarga la página HTML)
#   - BeautifulSoup → analiza el HTML para buscar y extraer datos
#   - pandas     → organiza los datos extraídos en una tabla (DataFrame)
# ============================================================

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# ============================================================
# STEP 1 – Download the first catalogue page
# English:
#   We build the URL for page 1 of the catalogue.
#   requests.get() downloads the raw HTML of that page.
#   We check the status code: 200 = success, anything else = problem.
#
# Español:
#   Construimos la URL de la página 1 del catálogo.
#   requests.get() descarga el HTML crudo de esa página.
#   Comprobamos el código de estado: 200 = éxito, otro valor = problema.
# ============================================================

base_url = "https://books.toscrape.com/catalogue/"
url = base_url + "page-1.html"

response = requests.get(url)
print("HTTP Status:", response.status_code)  # 200 = OK
```

```python
# ============================================================
# STEP 2 – Parse the HTML with BeautifulSoup
# English:
#   BeautifulSoup receives the raw bytes (response.content) and
#   the parser name ("html.parser" is built into Python, no install needed).
#   It creates a tree-like object we can search with methods like .find().
#
# Español:
#   BeautifulSoup recibe los bytes crudos (response.content) y
#   el nombre del parser ("html.parser" viene incluido en Python).
#   Crea un objeto en forma de árbol que podemos buscar con .find().
# ============================================================

soup = BeautifulSoup(response.content, "html.parser")

# Look at ONE book card to understand the HTML structure
# Observar UNA tarjeta de libro para entender la estructura HTML
first_book = soup.find("article", class_="product_pod")
print(first_book.prettify())
```

```python
# ============================================================
# HELPER 1 – Convert word rating to number
# English:
#   The website stores ratings as CSS class names: "One", "Two", ..., "Five".
#   We need to convert those words to integers (1, 2, 3, 4, 5) so we can
#   compare them with the min_rating parameter.
#
# Español:
#   El sitio almacena los ratings como clases CSS: "One", "Two", ..., "Five".
#   Necesitamos convertir esas palabras a enteros (1, 2, 3, 4, 5) para poder
#   compararlos con el parámetro min_rating.
# ============================================================

def word_to_rating(word):
    """
    Converts a word rating (e.g. 'Four') to an integer (e.g. 4).
    Convierte un rating en palabras (ej. 'Four') a un entero (ej. 4).
    """
    rating_map = {
        "One": 1,
        "Two": 2,
        "Three": 3,
        "Four": 4,
        "Five": 5
    }
    # .get() returns 0 if the word is not found, avoiding errors
    # .get() devuelve 0 si la palabra no existe, evitando errores
    return rating_map.get(word, 0)


# Quick test / Prueba rápida
print(word_to_rating("Four"))   # Expected: 4
print(word_to_rating("One"))    # Expected: 1
```

```python
# ============================================================
# HELPER 2 – Scrape the detail page of a single book
# English:
#   The listing page only shows title, price, rating, and availability.
#   Genre, UPC, and Description are only available on each book's own page.
#   This function downloads a book detail page and extracts those extra fields.
#
# Español:
#   La página de listado sólo muestra título, precio, rating y disponibilidad.
#   El género, UPC y descripción solo están en la página individual del libro.
#   Esta función descarga esa página y extrae esos campos adicionales.
# ============================================================

def get_book_details(detail_url):
    """
    Downloads a book's detail page and returns a dict with:
    UPC, Genre, Description.

    Descarga la página de detalle de un libro y devuelve un dict con:
    UPC, Género, Descripción.
    """
    response = requests.get(detail_url)
    soup = BeautifulSoup(response.content, "html.parser")

    # ---- UPC ----
    # English: The UPC is inside a <table> with class "table table-striped".
    #          Each row has <th> (label) and <td> (value). We look for "UPC".
    # Español: El UPC está dentro de una <table class="table table-striped">.
    #          Cada fila tiene <th> (etiqueta) y <td> (valor). Buscamos "UPC".
    upc = "N/A"  # default if not found / valor por defecto si no se encuentra
    table = soup.find("table", class_="table table-striped")
    if table:
        rows = table.find_all("tr")
        for row in rows:
            header = row.find("th")
            value  = row.find("td")
            if header and value and header.text.strip() == "UPC":
                upc = value.text.strip()
                break  # stop once found / detenemos cuando lo encontramos

    # ---- GENRE ----
    # English: The genre appears in the breadcrumb navigation at the top of the page.
    #          Breadcrumb looks like: Home > Mystery > [Book Title]
    #          Genre is the THIRD element (index 2) of the breadcrumb list.
    # Español: El género aparece en el breadcrumb (miga de pan) de la página.
    #          El breadcrumb es: Home > Misterio > [Título del libro]
    #          El género es el TERCER elemento (índice 2) de esa lista.
    genre = "N/A"
    breadcrumb = soup.find("ul", class_="breadcrumb")
    if breadcrumb:
        items = breadcrumb.find_all("li")
        if len(items) >= 3:
            genre = items[2].text.strip()  # index 0=Home, 1=Books, 2=Genre, 3=Title

    # ---- DESCRIPTION ----
    # English: The description is inside the first <p> tag that comes after
    #          the <div id="product_description"> element.
    #          Some books have no description, so we use a fallback.
    # Español: La descripción está dentro del primer <p> que sigue al
    #          elemento <div id="product_description">.
    #          Algunos libros no tienen descripción, así que usamos un valor por defecto.
    description = "No description available"
    desc_header = soup.find("div", id="product_description")
    if desc_header:
        next_p = desc_header.find_next_sibling("p")
        if next_p:
            description = next_p.text.strip()

    return {
        "UPC": upc,
        "Genre": genre,
        "Description": description
    }


# Quick test with a known book
# Prueba rápida con un libro conocido
test_details = get_book_details("https://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html")
print(test_details)
```

```python
# ============================================================
# MAIN FUNCTION – scrape_books
# English:
#   This function iterates through ALL pages of the Books to Scrape catalogue.
#   For each book on each page it:
#     1. Extracts title, price, rating, availability from the listing card.
#     2. Applies the min_rating and max_price filters immediately.
#        → If a book fails either filter, we SKIP it (no detail page request)
#          This avoids making hundreds of unnecessary HTTP requests.
#     3. Only visits the detail page for books that PASS both filters.
#     4. Collects all data into a list, then converts it to a DataFrame.
#
# Español:
#   Esta función recorre TODAS las páginas del catálogo de Books to Scrape.
#   Para cada libro en cada página:
#     1. Extrae título, precio, rating y disponibilidad de la tarjeta del listado.
#     2. Aplica los filtros min_rating y max_price de inmediato.
#        → Si un libro no pasa algún filtro, lo SALTAMOS (sin petición de detalle)
#          Esto evita cientos de peticiones HTTP innecesarias.
#     3. Sólo visita la página de detalle para libros que PASAN ambos filtros.
#     4. Recopila todos los datos en una lista y la convierte en DataFrame.
# ============================================================

def scrape_books(min_rating, max_price):
    """
    Scrapes books.toscrape.com and returns a DataFrame of books
    that satisfy both conditions:
      - Rating  >= min_rating  (integer 1-5)
      - Price   <= max_price   (float, in GBP £)

    Parameters / Parámetros:
      min_rating (int/float) : minimum star rating (1–5)
      max_price  (float)     : maximum price in pounds

    Returns / Devuelve:
      pandas.DataFrame with columns:
        UPC, Title, Price (£), Rating, Genre, Availability, Description
    """

    # ---------------------------------------------------------
    # Base URL for the catalogue pages
    # URL base para las páginas del catálogo
    # ---------------------------------------------------------
    base_url = "https://books.toscrape.com/catalogue/"

    # List that will hold one dict per book that passes the filters
    # Lista que guardará un dict por libro que pase los filtros
    books_data = []

    # Start from page 1 and keep going until there are no more pages
    # Empezamos desde la página 1 y continuamos hasta que no haya más
    page_number = 1

    while True:  # We'll break out of this loop when we run out of pages

        # ---- Build the page URL ----
        # Construir la URL de la página actual
        page_url = base_url + f"page-{page_number}.html"
        response = requests.get(page_url)

        # ---- Stop if page doesn't exist ----
        # English: A 404 status means the page doesn't exist → we've gone past the last page.
        # Español: Un código 404 significa que la página no existe → pasamos de la última.
        if response.status_code == 404:
            print(f"  → No more pages after page {page_number - 1}. Stopping.")
            break

        print(f"Scraping page {page_number}...")

        # Parse the HTML of the current listing page
        # Parsear el HTML de la página de listado actual
        soup = BeautifulSoup(response.content, "html.parser")

        # Find all book cards on this page
        # English: Every book is wrapped in an <article class="product_pod"> tag.
        # Español: Cada libro está envuelto en una etiqueta <article class="product_pod">.
        book_cards = soup.find_all("article", class_="product_pod")

        # ---- Loop through each book card on the page ----
        # Recorrer cada tarjeta de libro en la página
        for card in book_cards:

            # == EXTRACT RATING ==
            # English: The rating is a CSS class on the <p class="star-rating Four"> element.
            #          We grab the SECOND class (index 1) — the first is always "star-rating".
            # Español: El rating es una clase CSS en <p class="star-rating Four">.
            #          Tomamos la SEGUNDA clase (índice 1) — la primera siempre es "star-rating".
            rating_tag = card.find("p", class_="star-rating")
            rating_word = rating_tag["class"][1]   # e.g. "Four"
            rating = word_to_rating(rating_word)    # e.g. 4

            # == FILTER BY RATING (early exit) ==
            # English: If the rating is below our minimum, skip this book immediately.
            #          This avoids loading the detail page unnecessarily.
            # Español: Si el rating es menor que el mínimo, saltamos este libro.
            #          Así evitamos cargar la página de detalle innecesariamente.
            if rating < min_rating:
                continue  # go to next book / ir al siguiente libro

            # == EXTRACT PRICE ==
            # English: Price is inside <p class="price_color"> as text like "£12.50".
            #          We strip the "£" symbol and convert to float for comparison.
            # Español: El precio está en <p class="price_color"> como texto "£12.50".
            #          Quitamos el símbolo "£" y convertimos a float para comparar.
            price_text = card.find("p", class_="price_color").text.strip()
            price = float(price_text.replace("£", "").strip())

            # == FILTER BY PRICE (early exit) ==
            # If price exceeds maximum, skip this book
            # Si el precio supera el máximo, saltamos este libro
            if price > max_price:
                continue

            # == EXTRACT TITLE ==
            # English: Title is in the "title" attribute of the <a> inside <h3>.
            #          (The visible text is truncated; the attribute has the full title.)
            # Español: El título está en el atributo "title" del <a> dentro del <h3>.
            #          (El texto visible está truncado; el atributo tiene el título completo.)
            title = card.find("h3").find("a")["title"]

            # == EXTRACT AVAILABILITY ==
            # English: Availability text is inside <p class="instock availability">.
            #          We use .strip() to remove extra whitespace.
            # Español: La disponibilidad está en <p class="instock availability">.
            #          Usamos .strip() para eliminar espacios en blanco extra.
            availability_tag = card.find("p", class_="availability")
            availability = availability_tag.text.strip()

            # == BUILD DETAIL PAGE URL ==
            # English: The href in each card looks like "../../../some-book/index.html".
            #          We replace the "../../../" prefix with our base_url to get the full URL.
            # Español: El href en cada tarjeta tiene la forma "../../../some-book/index.html".
            #          Reemplazamos el prefijo "../../../" con base_url para obtener la URL completa.
            relative_link = card.find("h3").find("a")["href"]
            detail_url = base_url + relative_link.replace("../", "")

            # == FETCH DETAIL PAGE (UPC, Genre, Description) ==
            # Only called for books that passed BOTH filters
            # Solo se llama para libros que pasaron AMBOS filtros
            details = get_book_details(detail_url)

            # == APPEND BOOK RECORD ==
            # English: Build a dictionary with ALL required columns and add it to our list.
            # Español: Construimos un diccionario con TODAS las columnas requeridas y lo añadimos.
            books_data.append({
                "UPC":          details["UPC"],
                "Title":        title,
                "Price (£)":    price,
                "Rating":       rating,
                "Genre":        details["Genre"],
                "Availability": availability,
                "Description":  details["Description"]
            })

        # Move to the next page / Pasar a la siguiente página
        page_number += 1

    # ---- Build and return the DataFrame ----
    # English: Convert our list of dicts to a pandas DataFrame.
    #          reset_index(drop=True) gives clean sequential indices (0, 1, 2, ...).
    # Español: Convertimos nuestra lista de dicts en un DataFrame de pandas.
    #          reset_index(drop=True) da índices secuenciales limpios (0, 1, 2, ...).
    df = pd.DataFrame(books_data)
    df = df.reset_index(drop=True)
    return df
```

```python
# ============================================================
# RUN THE SCRAPER
# English:
#   We call scrape_books with the parameters from the lab instructions.
#   This will take a few minutes because it visits hundreds of pages.
#   You will see progress printed for each page.
#
# Español:
#   Llamamos a scrape_books con los parámetros indicados en el enunciado.
#   Tardará unos minutos porque visita cientos de páginas.
#   Verás el progreso impreso en cada página.
# ============================================================

df = scrape_books(min_rating=4, max_price=20)

print(f"\n✅ Done! Found {len(df)} books matching the criteria.")
print(f"✅ ¡Listo! Se encontraron {len(df)} libros que cumplen los criterios.")
```
